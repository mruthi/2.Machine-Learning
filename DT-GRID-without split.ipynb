```python
import sklearn 
```

```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid={'criterion':['mae','friedman_mse','mse',],'max_features':['auto','sqrt','log2'],'splitter':['best','random']}
grid=GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("The R score value for best parameter:{}".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("age:"))
bmi_input=float(input("BMI:"))
children_input=int(input("children:"))
sex_male_input=int(input("sex male 0 or 1:"))
smoker_yes_input=int(input("smoker yes 0 or 1:"))
```

```python
Future_predicion=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future Prediction: {}".format(Future_predicion))
```
