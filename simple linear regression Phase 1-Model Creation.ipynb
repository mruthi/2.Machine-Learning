```python
#pandas is a library with all functions.we can import functions from here.we can mention pandas shortly as pd
import pandas as pd#if we import pandas all the functions of pandas will be stored in this file.here we used pandas for table
```

```python
#dataset is a variable name .for this variable we r assigning dataset by using pd class with read_csv fn
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#here we are getting input data column separately by using independent variable name 
independent=dataset[["YearsExperience"]]
```

```python
#In model creation phase  we splitted the input and output 
independent
```

```python
#ere we are getting output data column separately by using dependent variable name
dependent=dataset[["Salary"]]
```

```python
dependent
```

```python
#here we r going to split the traing set and test set by using sklearn library.
#For memory saving we import particular fn from sklearn library
from sklearn.model_selection import train_test_split
                               #(passing input and output parameters)
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
#we are assigning 4 variables ,(x means input and y means output)for train set and tes set
```

```python
#sklearn library having all fucntion for ML.
#For Regression we are creating regressor variable name.
from sklearn.linear_model import LinearRegression
# Now all the functions of LinearRegression class will be assign to the regressor variable
regressor=LinearRegression()
#the first two lines are procedure
regressor.fit(x_train,y_train)
#here we substitute the training inputs by using fit method.
#once passed the input values we will get the weight and bais.so the model creation is done
```

```python
#coef is the method to calculate the weight.
#we are checking the value of weight here
weight=regressor.coef_
weight
```

```python
##intercept is the method to calculate the bais or origin.
bais=regressor.intercept_
bais
```

```python
#Evaluvation metrics process for test data
#here we cross check the prediction value with  our brain created output
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
#y_test is a actual output,y_pred is predicted output 
#here we are comparing the both by usimg r2_score fn
r_score=r2_score(y_test,y_pred)
```

```python
#the value nearly 1 .so best model 
r_score
```

```python
#model saving process by using pickle library
#creating file name for save this model
#.sav is pickle extension
import pickle
filename="finalized_model_linear.sav"
```

```python
#Here we are insert the regressor file to the filename 
pickle.dump(regressor,open(filename,'wb'))
```

```python
#loading the saved model
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[0]])
```
